```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('/Users/katerynakovalchuk/Downloads/Coffee Shop Sales.xlsx')
df.head()
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.info()
```

```python
df['transaction_time'] = df['transaction_time'].astype(str)
df['transaction_time'] = pd.to_timedelta(df['transaction_time'])
```

```python
df['sales']=df['transaction_qty']*df['unit_price']
```

```python
df['datetime']=df['transaction_date']+df['transaction_time']

df.head()
```

```python
df.describe()
```

```python
df_location = df.groupby('store_location').agg({
    'sales': 'sum',
    'transaction_id': 'count'
})

df_location
```

```python
daily_sales_by_location = df.groupby(['transaction_date','store_location'])['sales'].sum().unstack()

daily_sales_by_location
```

```python
test = df.groupby(['transaction_date','store_location'])['sales'].sum()

test
```

```python
daily_sales_by_location.plot(figsize=(14,8), title='Daily Sales by Store Location')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(14,8))

plt.stackplot(daily_sales_by_location.index, daily_sales_by_location.T, labels=daily_sales_by_location.columns)

plt.title('Daily Sales by Store Location (Stacked Area Chart)')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.legend(title='Store Location', loc = 'upper left')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
daily_sales = df.groupby('transaction_date')['sales'].sum().reset_index()

df['day_of_week'] = df['datetime'].dt.day_name()

weekly_sales = df.groupby('day_of_week')['sales'].sum().reindex(
    ['Monday','Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']).reset_index()

df['month'] = df['datetime'].dt.to_period('M')
monthly_sales = df.groupby('month')['sales'].sum().reset_index()

daily_sales, weekly_sales, monthly_sales
```

```python
monthly_sales.plot(figsize=(14,8), title='mMothly Sales')

plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
weekly_sales = df.groupby(['day_of_week','store_location'])['sales'].sum().unstack()

days_order = ['Monday','Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales = weekly_sales.reindex(days_order)
weekly_sales
```

```python
weekly_sales[["Astoria", "Hell's Kitchen", "Lower Manhattan"]].plot(kind='bar',figsize=(14,8))
plt.xlabel('Day of the week', fontweight = 'bold')
plt.ylabel('Total Sales ($)', fontweight = 'bold')
plt.title ('Weekly Sales by Store Location')
plt.legend(title = 'Store Location', loc='upper left')

plt.tight_layout()
plt.show()
```

```python
df_category = df.groupby('product_category').agg({
    'sales': 'sum',
    'transaction_id': 'count'
}).sort_values('sales', ascending = False)

df_category
```

```python
df['hour'] = df['datetime'].dt.hour

hourly_sales_by_product = df.groupby(['hour','product_category'])['sales'].sum().unstack()

hourly_sales_by_product.plot(kind='bar',stacked=True, figsize=(14,8),colormap='viridis')

plt.title('Hourly Sales by Product Category')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Category', loc='upper right')
plt.grid(True)
plt.tight_layout()

plt.show()
```

```python
df_category['avg_sales'] = df_category['sales']/df_category['transaction_id']
df_category
```

```python
sales_by_product = df.groupby('product_type').agg({
    'sales': 'sum',
    'unit_price': 'mean'
}).sort_values('sales', ascending = False)

sales_by_product
```

```python
product_sales_df = df.groupby(['product_category', 'product_type'])['sales'].sum().reset_index()
product_sales_df
```

```python
plt.figure(figsize=(14,8))
sns.barplot(data=product_sales_df, x='product_category', y='sales', hue='product_type')

plt.title('Product Sales by Category and Type')
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Type', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
categories = product_sales_df['product_category'].unique()

for category in categories:
    plt.figure(figsize=(5,5))
    category_data = product_sales_df[product_sales_df['product_category']==category]
    sns.barplot(data=category_data,x='product_type',y='sales')
    
    plt.title(f'Sales by Product Type in {category}')
    plt.xlabel('Product Type')
    plt.ylabel('Total Sales ($)')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
```

```python
hourly_sales_by_location = df.groupby(['store_location','hour'])['sales'].sum().reset_index()
hourly_sales_by_location
```

```python
df['time_of_day'] = pd.cut(df['hour'], bins=[0, 8, 12, 18, 24], 
                            labels=['Morning', 'Afternoon', 'Evening', 'Night'])

popularity_by_time = df.groupby(['time_of_day', 'product_category'])['sales'].sum().unstack()

popularity_by_time.plot(kind='bar', figsize=(14, 8))
plt.title('Popularity of Products by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Category')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
location_sales = df.groupby('store_location').agg({
    'sales': 'sum',
    'transaction_id': 'count'
})

location_sales['sales'].plot(kind='bar', figsize=(14, 8))
plt.title('Total Sales by Store Location')
plt.xlabel('Store Location')
plt.ylabel('Total Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
avg_check = df.groupby('store_location')['sales'].mean()
total_transactions = df.groupby('store_location')['transaction_id'].count()

avg_check_df = pd.DataFrame({'Average Check': avg_check, 'Total Transactions': total_transactions})

avg_check_df.plot(kind='bar', figsize=(14, 8), secondary_y='Total Transactions')
plt.title('Average Check and Total Transactions by Store Location')
plt.xlabel('Store Location')
plt.ylabel('Average Check ($)')
plt.right_ylabel('Total Transactions')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_category = df.groupby('product_category')['sales'].sum().reset_index()
df_category['Cumulative Sales'] = df_category['sales'].cumsum()
df_category['Sales Share'] = df_category['sales'] / df_category['sales'].sum()
df_category['Cumulative Share'] = df_category['Sales Share'].cumsum()

df_category['Category'] = pd.cut(df_category['Cumulative Share'], bins=[0, 0.8, 0.95, 1.0], 
                                  labels=['A', 'B', 'C'])

sns.barplot(data=df_category, x='product_category', y='sales', hue='Category')
plt.title('ABC Analysis of Product Categories')
plt.xlabel('Product Category')
plt.ylabel('Total Sales ($)')
plt.legend(title='Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Создание складного графика для категорий продуктов
categories = df['product_category'].unique()

for category in categories:
    plt.figure(figsize=(14, 8))
    category_data = df[df['product_category'] == category]
    trends = category_data.groupby(['month', 'product_detail'])['sales'].sum().unstack()

    trends.plot(kind='line')
    plt.title(f'Sales Trends for {category}')
    plt.xlabel('Month')
    plt.ylabel('Total Sales ($)')
    plt.legend(title='product_detail')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
top_products = df.groupby('product_detail')['sales'].sum().nlargest(10).index
top_product_trends = df[df['product_detail'].isin(top_products)]

top_product_trends = top_product_trends.groupby(['month', 'product_detail'])['sales'].sum().unstack()

top_product_trends.plot(kind='line', figsize=(14, 8))
plt.title('Sales Trends for Top 10 Products')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.legend(title='Product Detail')  
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['day_of_week'] = df['transaction_date'].dt.day_name()

ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=ordered_days, ordered=True)

frequency_by_time = df.groupby(['day_of_week', 'hour'])['sales'].sum().unstack()  # Используем 'sales' вместо 'transaction_id'

plt.figure(figsize=(14, 8))
sns.heatmap(frequency_by_time, cmap='YlGnBu', annot=True, fmt='.2f')  # fmt='.2f' для отображения с двумя десятичными
plt.title('Total Sales by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.tight_layout()
plt.show()
```

```python
df['day_of_week'] = df['transaction_date'].dt.day_name()

ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=ordered_days, ordered=True)

frequency_by_time = df.groupby(['day_of_week', 'hour'])['transaction_id'].count().unstack()

plt.figure(figsize=(14, 8))
sns.heatmap(frequency_by_time, cmap='YlGnBu', annot=True, fmt='g')  # Изменяем fmt на 'g' для обычного формата
plt.title('Frequency of Purchases by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.tight_layout()
plt.show()
```

```python
df['day_of_week'] = df['transaction_date'].dt.day_name()

ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_of_week'] = pd.Categorical(df['day_of_week'], categories=ordered_days, ordered=True)


sales_summary = df.groupby(['day_of_week', 'hour']).agg(total_sales=('sales', 'sum'), transaction_count=('transaction_id', 'count'))

sales_summary['average_check'] = sales_summary['total_sales'] / sales_summary['transaction_count']

average_check_by_time = sales_summary['average_check'].unstack()

plt.figure(figsize=(14, 8))
sns.heatmap(average_check_by_time, cmap='YlGnBu', annot=True, fmt='.2f')  # fmt='.2f' для отображения с двумя десятичными
plt.title('Average Check by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.tight_layout()
plt.show()
```
